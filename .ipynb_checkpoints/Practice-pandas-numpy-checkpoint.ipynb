### Dot Product Example

```python
np.random.seed(0)
#Numbers of Jars Sold
sales_amt = np.random.randint(20,size=(5,3))
sales_amt
```

```python
#Creates weekly sales 
weekly_sales = pd.DataFrame(sales_amt,index=["Mon","Tues","Wed","Thrus","Fri"],
                           columns=["Almond Butter","Peanut Butter","Cashew Butter"])
weekly_sales
```

```python
#Create Prices array
prices = np.array([10,8,12])
prices
```

```python
#Create the dataframme
butter_prices = pd.DataFrame(prices.reshape(1,3),index=["Price"],columns=["Almond Butter","Peanut Butter","Cashew Butter"])
butter_prices
```

```python
#The sales_amt is not trasposed lets transpose it
total_sales = prices.dot(sales_amt.T)
total_sales
```

```python
butter_prices.shape , weekly_sales.shape
```

```python
#Create Daily sales
daily_sales = butter_prices.dot(weekly_sales.T)
daily_sales
```

```python
#Addinf daily sales column in weekly_sales dataframe
weekly_sales["Total $"] = daily_sales.T
weekly_sales
```

### When to use pyplot and when to use the OO method
* When plotting something quickly ,use the pyplot method
* When plotting something  more advvanced use the OO method

```python
over_50 = heart_disease[heart_disease["age"]>50]
over_50
```

```python
#pyplot method
over_50.plot(kind="scatter",
            x="age",
            y="chol",
            c="target");
```

```python
#matplotlib OO method
fix,ax = plt.subplots(figsize=(10,6))
over_50.plot(kind="scatter",
            x="age",
            y="chol",
            c="target",
            ax=ax);
ax.set_xlim([45,100])
```

```python
#OO method from scratch
fix,ax = plt.subplots(figsize=(10,6))
#Plot the data 
scatter = ax.scatter(x=over_50["age"],
                     y=over_50["chol"],
                     c=over_50["target"])
#Customize the plot
ax.set(title="Heart Disease and Cholestrol Levels",
      xlabel="Age",
      ylabel="Cholesterol")
#Legend
ax.legend(*scatter.legend_elements(),title="Target")
#add a horizontal Line
ax.axhline(over_50["chol"].mean(),linestyle="--")
```

```python
#OO method from scratch
fig,(ax0,ax1) = plt.subplots(nrows=2,
                             ncols=1,
                             figsize=(10,10))
#Plot the data ax0
scatter = ax0.scatter(x=over_50["age"],
                     y=over_50["chol"],
                     c=over_50["target"])
#ax1
scatter1 = ax1.scatter(x=over_50["age"],
                     y=over_50["thalach"],
                     c=(over_50["target"]))
                
#Customize the plot ax0
ax0.set(title="Heart Disease and Cholestrol Levels",
      xlabel="Age",
      ylabel="Cholesterol")
#ax1
ax1.set(title="Heart Disease and Thalach Levels",
      xlabel="Age",
      ylabel="Cholesterol")
#Legend ax0
ax0.legend(*scatter.legend_elements(),title="Target")
#ax1
ax1.legend(*scatter.legend_elements(),title="Target")
#add a horizontal Line ax0
ax0.axhline(over_50["chol"].mean(),linestyle="--")
#ax1
ax1.axhline(over_50["thalach"].mean(),linestyle="--")

#adding title to the graph
fig.suptitle("Heart Disease Analysis",fontsize=20,fontweight="bold");
```
